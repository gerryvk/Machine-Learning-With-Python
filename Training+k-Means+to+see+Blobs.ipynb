```python
#import the algorithm
from sklearn.cluster import KMeans
#import the dataset
from sklearn.datasets.samples_generator import make_blobs
import numpy as np
#stuff for displaying graphs
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
%matplotlib inline
```

```python
#create a 'blobs' dataset with 6 blobs using 300 samples
X, y = make_blobs(n_samples=300, centers=6, cluster_std=0.60, random_state=0)
```

```python
#examin what was created
print("X is a {},".format(type(X)), "shape is (samples, features): {}\n".format(X.shape))
print("y is a one dimensional array with it's associated cluster number:\n", y)
```

```python
#Plot blobs dataset to expose the blobs
plt.scatter(X[:, 0], X[:, 1], s=20);
```

```python
#teach KMeans using the X dataset
kmeans = KMeans(n_clusters=6) #encapsulate the algorithm in 'kmeans' & tell it how many clusters there are (see above)
kmeans.fit(X) #load the data contained in X
y_kmeans = kmeans.predict(X) #make a prediction
```

```python
#visualize the results plotting the data with coloured labels
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20, cmap='viridis');
```

```python
#visualize the results and the cluster centers as determined by the k-means estimator
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=20, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=150);
```
